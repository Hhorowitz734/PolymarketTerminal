# Display Timeseries Data from TS Script

```python
import json
import sys
import matplotlib.pyplot as plt
from datetime import datetime

def plot_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

    # Assuming the data structure is something like: { "data": [ { "t": timestamp, "p": price }, ... ] }
    times = [point["t"] for point in data["history"]]
    prices = [point["p"] for point in data["history"]]

    # Convert UNIX timestamps to readable dates
    times_readable = [datetime.fromtimestamp(t) for t in times]

    plt.figure(figsize=(10, 5))
    plt.plot(times_readable, prices, marker='o')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.title(f"Price history from {file_path}")
    plt.grid(True)
    plt.show()

```

```python
plot_json("timeseries_ts/yes_prices.json")
```
